```python
right = 0
wrong = 0
length = int(input('How many rounds'))
for i in range(length):
    x = random.randrange(1,10,1)
    y = random.randrange(1,10,1)
    solution = x*y
    print(f'evaluate {x} * {y}')
    answer = int(input())
    if answer == solution:
        print('Yay')
        right += 1
    else: 
        print('Wrong')
        wrong += 1
print(f'you got {right}/{length}')
```
