# Poynomials

## Linear Functions
Polynomical function of degree 1 or less.   
  
$$
f(x) = ax +b
$$

Where $a$ and $b$ are constants. The result is a graph of a straight line. $a$ is referred to as the slope, and $b$ is the intercept.  

Let us set $a=2$ and $b=2$  
$$
f(x) = 2x+2
$$

```python
def linear(x, a, b):
    return a*x + b

x = np.arange(-5, 5,.1)
fig, ax = plt.subplots()
ax.plot(x, linear(x,2,2))
ax.set_aspect('equal')
ax.grid(True, which='both')
ax.set_ylim(-5,5)
# set the x-spine (see below for more info on `set_position`)
ax.spines['left'].set_position('zero')

# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

# set the y-spine
ax.spines['bottom'].set_position('zero')

# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()
```

### Derivitive of linear functions:

Since the derivitive of the function defines the rate of change of a function, and linear functions have constant rates of change, the derrivitve of these functions is a constant.  
  
$$
\begin{align}
f(x) = ax + b\\
f'(x) = a
\end{align}
$$

For the example:  

$$
f'(x) = 2
$$


```python
def dlinear(x, a):
    return a

fig, ax = plt.subplots()
ax.plot(x, linear(x,2,2))
ax.axhline(y=2)

ax.set_aspect('equal')
ax.grid(True, which='both')
ax.set_ylim(-5,5)
ax.set_xlim(-5,5)
# set the x-spine (see below for more info on `set_position`)
ax.spines['left'].set_position('zero')

# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

# set the y-spine
ax.spines['bottom'].set_position('zero')

# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()
```

### Integral of Linear Functions

The integral descibes the area under the "curve". In the case of linear functions 

$$
\begin{align}
f(x) = 2x+2 \\
F(x) = \int{f(x)}dx \\
F(x) = \int{2x+2}dx \\
F(x) = \frac{1}{2}2x^{2} + 2x + c
\end{align}
$$

Where c is the constant of integration  

Graphing the integral would be graphing the accumulated amount of area under the curve.  


The graph would then show the amount of area under the linear function at each x value.   


Since this is dependent on the postion of x, we shall graph the definite integral over the range (-5, 5) and assume a constant of integration to be 0. 

```python
def intlinear(x, a, b, c):
    return (1/2)*a*x**(2) + b*x + c

fig, ax = plt.subplots()
ax.plot(x, linear(x,2,2))
ax.plot(x, intlinear(x, 2, 2, -10))

ax.set_aspect('equal')
ax.grid(True, which='both')
ax.set_ylim(-5,5)
ax.set_xlim(-5,5)
# set the x-spine (see below for more info on `set_position`)
ax.spines['left'].set_position('zero')

# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

# set the y-spine
ax.spines['bottom'].set_position('zero')

# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()
```

## Quadratic Functions

Polynomial functions of degree 2 

$$
f(x) = ax^{2} + bx + c
$$


```python
def quadratic(x, a, b, c):
    return a*x**2 + b*x + c
fig, ax = plt.subplots()

ax.plot(x, quadratic(x, 1, 0, 0))

ax.set_aspect('equal')
ax.grid(True, which='both')
ax.set_ylim(-5,5)
ax.set_xlim(-5,5)
# set the x-spine (see below for more info on `set_position`)
ax.spines['left'].set_position('zero')

# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

# set the y-spine
ax.spines['bottom'].set_position('zero')

# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()
```

### Derrivitive of quadratic function:

The derivitive of a quadratic function is a linear function.

$$
\begin{align}
f(x) = ax^{2} + bx + c\\
f'(x) = 2*ax + b
\end{align}
$$

```python
fig, ax = plt.subplots()

ax.plot(x, quadratic(x, 1, 0, 0))
ax.plot(x, linear(x, 1, 0))
ax.set_aspect('equal')
ax.grid(True, which='both')
ax.set_ylim(-5,5)
ax.set_xlim(-5,5)
# set the x-spine (see below for more info on `set_position`)
ax.spines['left'].set_position('zero')

# turn off the right spine/ticks
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()

# set the y-spine
ax.spines['bottom'].set_position('zero')

# turn off the top spine/ticks
ax.spines['top'].set_color('none')
ax.xaxis.tick_bottom()
```
